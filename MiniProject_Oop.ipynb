```python
#EX1
#Tache 1
class Livre:
    def __init__(self, titre, auteur, annee):
        self.titre = titre
        self.auteur = auteur
        self.annee = annee
        self.disponible = True

    def emprunter(self):
        if self.disponible:
            self.disponible = False
            print(f"{self.titre} emprunté avec succès!")
        else:
            print(f"Erreur: {self.titre} n'est pas disponible.")

    def rendre(self):
        if not self.disponible:
            self.disponible = True
            print(f"{self.titre} rendu avec succès!")
        else:
            print(f"Erreur: {self.titre} était déjà disponible.")

    def __str__(self):
        statut = "Disponible" if self.disponible else "Emprunté"
        return f"{self.titre} — {self.auteur} ({self.annee}) | {statut}"
```

```python
#Tache2
class Bibliotheque:
    def __init__(self):
        self.livres = []

    def ajouter_livre(self, livre):
        self.livres.append(livre)
        print(f"Livre '{livre.titre}' ajouté à la bibliothèque.")

    def afficher_livres(self):
        if not self.livres:
            print("La bibliothèque est vide.")
            return
        print("\n=== CATALOGUE ===")
        for i, livre in enumerate(self.livres, 1):
            print(f"{i}. {livre}")

    def rechercher_par_titre(self, mot):
        trouve = False
        print(f"\nRecherche pour '{mot}':")
        for livre in self.livres:
            if mot.lower() in livre.titre.lower():
                print(f"- {livre}")
                trouve = True
        if not trouve:
            print("Aucun livre trouvé.")

    def supprimer_livre(self, titre):
        for i, livre in enumerate(self.livres):
            if livre.titre.lower() == titre.lower():
                del self.livres[i]
                print(f"Livre '{titre}' supprimé.")
                return
        print(f"Livre '{titre}' non trouvé.")

    def emprunter_livre(self, titre):
        for livre in self.livres:
            if livre.titre.lower() == titre.lower():
                livre.emprunter()
                return
        print(f"Livre '{titre}' non trouvé.")

    def rendre_livre(self, titre):
        for livre in self.livres:
            if livre.titre.lower() == titre.lower():
                livre.rendre()
                return
        print(f"Livre '{titre}' non trouvé.")
```

```python
# Programme principal avec exemples de littérature française
biblio = Bibliotheque()

# Création de livres classiques français
l1 = Livre("Les Misérables", "Victor Hugo", 1862)
l2 = Livre("Notre-Dame de Paris", "Victor Hugo", 1831)
l3 = Livre("Le Petit Prince", "Antoine de Saint-Exupéry", 1943)
l4 = Livre("L'Étranger", "Albert Camus", 1942)
l5 = Livre("Voyage au bout de la nuit", "Louis-Ferdinand Céline", 1932)

# Ajout à la bibliothèque
biblio.ajouter_livre(l1)
biblio.ajouter_livre(l2)
biblio.ajouter_livre(l3)
biblio.ajouter_livre(l4)
biblio.ajouter_livre(l5)

# Afficher catalogue initial
biblio.afficher_livres()

# Emprunter un livre
biblio.emprunter_livre("Petit prince")
biblio.emprunter_livre("Notre-Dame de Paris")

# Afficher après emprunt
biblio.afficher_livres()

# Rendre un livre
biblio.rendre_livre("Petit Prince")

# Rechercher
biblio.rechercher_par_titre("voyage")

# Supprimer un livre
biblio.supprimer_livre("L'Étranger")

# Afficher catalogue final
biblio.afficher_livres()
```
